```python
import glob
import os
import json

import pandas as pd
import seaborn as sns


def read_json(x):
    with open(x) as xfp:
        j = json.load(xfp)
    jj = {}
    for k in j:
        jj[k.replace("mean_", "")] = j[k]
    jj["video"] = os.path.splitext(os.path.basename(x))[0]
    return jj

df = pd.DataFrame(map(read_json, glob.glob("features/*.json")))
df
```

```python
dmos = pd.read_csv("mos.csv")
dmos
```

```python
dm = pd.merge(dmos, df, left_on="vid", right_on="video")
dm
```

```python
values = []
for col in df.columns.difference(["video", "clip_genre_max"]):
    
    values.append(
        {
            "feature/model": col,
            "pearson" :dm[[col, "MOS full"]].corr()[col].iloc[-1],
            "spearman" :dm[[col, "MOS full"]].corr(method="spearman")[col].iloc[-1],
            "kendall" :dm[[col, "MOS full"]].corr(method="kendall")[col].iloc[-1],
        }
    )
    #break
dcorr = pd.DataFrame(values).sort_values(by="pearson")
dcorr
```

```python
sns.set(rc={'figure.figsize':(4,8)})
sns.set_style("white")

sns.barplot(x="pearson", y="feature/model", data=dcorr)
```

```python
values = []
for col in [x for x in dm.columns if "_cc" in x]:
    print(dm[[col, col.replace("_cc", "")]].corr())
    values.append(
        {
            "feature/model": col.replace("_cc", " vs. cc"),
            "pearson": dm[[col, col.replace("_cc", "")]].corr()[col].iloc[-1],
            "spearman":dm[[col, col.replace("_cc", "")]].corr(method="spearman")[col].iloc[-1],
            "kendall": dm[[col, col.replace("_cc", "")]].corr(method="kendall")[col].iloc[-1],
        }
    )

dcorr_cc = pd.DataFrame(values).sort_values(by="pearson")
dcorr_cc
```

```python
sns.set(rc={'figure.figsize':(10,1.5)})
sns.set_style("white")

ax = sns.heatmap(dcorr_cc.set_index("feature/model").transpose(), linewidth=.5, annot=True, fmt=".1f", cmap="crest", cbar=False)
ax.set_title("Heatmap of correlation values of \n features/models with their center cropped counter parts")
ax.get_figure().savefig("feature_model_corr_cc_heatmap.pdf", bbox_inches="tight")
```

```python
dfg = pd.read_json("time_feature_group.json")
dfg
```

```python
vv = []
for i, g in dfg.groupby(by="what"):
    for x in g["values"].values[0]:
        vv.append({"what": i, "time": x})
dfgv = pd.DataFrame(vv)
dfgv
```

```python
sns.boxplot(x="time", y="what", data=dfgv[dfgv["what"] != "all"][dfgv["what"] != "sequential"])
```

```python
all = dfgv[dfgv["what"]=="all"]["time"].mean()
```

```python
print(
dfgv.groupby(by="what").apply(
    lambda x: 
        pd.Series(
            {
                "mean_time": x.mean()[0],
                "std_time": x.std()[0],
                "%_all": 100 * x.mean()[0] / all
            }
        )
).sort_values(by="mean_time").reset_index().round(2).to_latex(index=False)
)
```
